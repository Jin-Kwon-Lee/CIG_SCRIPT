```python
# input_df = input_df.replace(" ",np.nan).dropna()
input_df = input_df.dropna()
col = input_df.columns
col[0]
input_df = input_df.set_index(col[0])
input_df.reset_index(inplace=True)
input_df.head(35)
```

```python
def _get_dict_description_info(dic,descript_list,BL_cnt):
    if type(descript_list) == list:
        descript_cate = descript_list[0]
        descript_val = descript_list[1]
        dic.update({BL_cnt:{descript_cate:descript_val}})
    elif type(descript_list) == str:
#         print(descript_list)
        cha_name = descript_list.split()[0]
        cha_year = descript_list.split()[1]
        cha_no = descript_list.split()[2]
#         print(cha_name,cha_year,cha_no)
        dic.update({BL_cnt:{'MODEL':cha_name,'YR':cha_year,'CHASSINO':cha_no}})
    return dic
```

```python
# shipper_cnt = 0
# shipper_start_idx = 1048000 #excel last cell index
BL_cnt = 0
con_dict = {}
chassi_dict = {}
acid_dict = {}
import_dict = {}
acid_dict = {}
exp_dict = {}
exp_cty_dict = {}

for idx,row in input_df.iterrows():
    val = row[0]
    
    if 'CONSIGNEE' in val:
        BL_cnt = BL_cnt + 1
        con_idx = idx
        con_name = val.split(':')
#         print(con_name)
        if idx >= 1:
            chassi_no_idx = idx - 1
            chassi_no = input_df.loc[chassi_no_idx,:][0]
            chassi_dict = _get_dict_description_info(chassi_dict,chassi_no,BL_cnt)
        con_dict = _get_dict_description_info(con_dict,con_name,BL_cnt)
    elif 'ACID' in val:
        acid_idx = idx
        acid_no = val.split(':')
        acid_dict = _get_dict_description_info(acid_dict,acid_no,BL_cnt)
#         print(acid_no)
    elif 'IMPORTER TAX NUMBER' in val:
        importer_idx = idx
        importer_no = val.split(':')
        import_dict = _get_dict_description_info(import_dict,importer_no,BL_cnt)
#         print(importer_no)
    elif 'FREIGHT FORWARDER ID' in val:
        exporter_idx = idx
        exporter_no = val.split(':')
        exp_dict = _get_dict_description_info(exp_dict,exporter_no,BL_cnt)
#         print(exporter_no)
    elif 'EXPORTER REGISTRATION COUNTRY' in val:
        exp_cty_idx = idx
        exp_cty_name = val.split(':')
        exp_cty_dict = _get_dict_description_info(exp_cty_dict,exp_cty_name,BL_cnt)
#         print(exp_cty_name)

print(chassi_dict)
print(con_dict)
print(acid_dict)
print(import_dict)
print(exp_dict)
print(exp_cty_dict)


#     total_dict.udpate(con_dict)
#     print(total_dict)



# shipper_start_idx, shipper_end_idx
```
