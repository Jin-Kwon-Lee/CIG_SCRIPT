```python
from tkinter import *
import tkinter as tk 
pd.options.display.max_rows = 50

def _check_if_empty_df(df):
    return df.empty

def _call_error_message_window(err_dict):
    window = Tk()
    window.title("This error is caused if empty location was found!")
    
#    Error message 
    lb_err_msg1 = Label(window, text="This error is caused if empty location in input file was found!")
    lb_err_msg1.place(x=10, y=10)
    
    lb_err_msg2 = Label(window, text="You should check below location of empty data in input file!")
    lb_err_msg2.place(x=10, y=30)
    
    #    Current working directory path
    working_dir_path = os.getcwd() + '\data'

    lb_err_working_dir = Label(window, text="Working Directory Path : ")
    lb_err_working_dir.place(x=10, y=70)
    
    ety_working_path = tk.Entry(fg="gray19", bg="snow", width=50)
    ety_working_path.place(x=10,y=90)
    ety_working_path.insert(0,working_dir_path)
    
    # label,entry location config
    default_y_row = 90
    lb_x_step_size = 0
    lb_y_step_size = 30
    lb_x_row = 10
    
    ety_x_step_size = 0
    ety_y_step_size = 60
    ety_x_row = 10
    
    for err_cnt in err_dict:
        empty_rows,empty_col = err_dict[err_cnt]
        
        #    N th Error Row Location
        lb_y_row = default_y_row + lb_y_step_size

        lb_err_empty_row = Label(window, text= "Empty Row Location : ")
        lb_err_empty_row.place(x=lb_x_row,y=lb_y_row)

        ety_y_row = default_y_row + ety_y_step_size 

        ety_empty_row = tk.Entry(fg="gray19", bg="snow", width=30)
        ety_empty_row.place(x=ety_x_row,y=ety_y_row)
        ety_empty_row.insert(0,empty_rows)

        #  N th  Error Column Location
        lb_err_empty_col = Label(window, text= "Empty Column Location : ")
        lb_err_empty_col.place(x=250, y=lb_y_row)
        ety_empty_col = tk.Entry(fg="gray19", bg="snow", width=30)
        ety_empty_col.place(x=250,y=ety_y_row)
        ety_empty_col.insert(0,empty_col)

        default_y_row = ety_y_row

    window.geometry('800x800')
    window.mainloop()
    

def _check_empty_col(df):
    df.replace('', np.nan, inplace=True)
    row_index = 2
    err_cnt = 0
    err_dict = {}
    
    for col in df.columns:
#         print(col)
#         print(df[col].isnull())
        check_each_col_isnull_df = df[col].isnull()
        
        if check_each_col_isnull_df.any():
            err_cnt = err_cnt + 1
            empty_index = df[check_each_col_isnull_df].index
            empty_row = list(empty_index + row_index)
            empty_col = col
            err_dict.update({err_cnt:(empty_row,empty_col)}) 
        else:
            pass
    
    for col in df.columns:
        check_each_col_isnull_df = df[col].isnull()
        if check_each_col_isnull_df.any():
            _call_error_message_window(err_dict)
            break
                
            
def check_valid_df(df):
    if _check_if_empty_df(df):
        warning_message()
    else:
        _check_empty_col(df)
```
